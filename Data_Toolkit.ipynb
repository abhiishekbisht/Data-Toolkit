```python
#Ques.4  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap.

import seaborn as sbn
import matplotlib.pyplot as plt
import pandas as pd
data = pd.DataFrame({'A': [1, 2, 3, 5, 7],
                     'B': [2, 3, 5, 6, 8],
                     'C': [5, 3, 6, 9, 10]})
correlation_matrix = data.corr()

sbn.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Ques.8 Create a line plot with multiple lines using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y1 = np.sin(x)               
y2 = np.cos(x)               
y3 = np.tan(x) 
plt.figure(figsize=(10, 5))  

plt.plot(x, y1, label='Sine', color='blue')
plt.plot(x, y2, label='Cosine', color='orange') 
plt.plot(x, y3, label='Tangent', color='green') 

plt.legend()
plt.title("Multiple Lines Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#Ques.9 Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold

import pandas as pd
data = {
    'Name' : ['A','B','C','D'],
    'Age' : [19,20,21,22],
    'Salary' : [50000, 40000, 35000, 42000]
}
df = pd.DataFrame(data)
threshold = 40000
filter_df = df[df['Salary'] > threshold]
print(filter_df)
```

```python
#Ques.10  Create a histogram using Seaborn to visualize a distribution.

import seaborn as sbn
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(loc=0, scale=1, size=1000)
plt.figure(figsize=(10, 6))  
sbn.histplot(data, bins=10, kde=True, color='orange', alpha=0.3)

plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

```python
#Ques.13 Create a 3D scatter plot using Plotly.

import plotly.express as px
import numpy as np


np.random.seed(42)
n_points = 100
x = np.random.rand(n_points) * 100  
y = np.random.rand(n_points) * 100  
z = np.random.rand(n_points) * 100  

data = {
    'X': x,
    'Y': y,
    'Z': z
}

fig = px.scatter_3d(data, x='X', y='Y', z='Z', title='3D Scatter Plot', 
                     labels={'X': 'X Axis', 'Y': 'Y Axis', 'Z': 'Z Axis'},
                     color=np.random.choice(['red', 'blue', 'green'], n_points))  # Random colors

fig.show()
```
